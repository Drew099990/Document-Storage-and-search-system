## Load and Preprocess Data

### Subtask:
Load the historical data including environmental factors and past patient numbers. Preprocess the data for PyTorch, handling missing values, scaling, and splitting into training, validation, and test sets. Define 'malaria outbreaks' as the target variable.


**Reasoning**:
The first step is to load the historical data into a pandas DataFrame, inspect its structure, data types, and descriptive statistics to understand the dataset's basic characteristics. I will assume the data is in a CSV file named 'historical_malaria_data.csv'.



```python
import pandas as pd

# Load the historical data
# Assuming the data is in a CSV file named 'historical_malaria_data.csv'
try:
    df = pd.read_csv('historical_malaria_data.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: 'historical_malaria_data.csv' not found. Please upload the file or specify the correct path.")
    # Create a dummy DataFrame for demonstration if the file is not found
    data = {
        'Date': pd.to_datetime(pd.date_range(start='2010-01-01', periods=100)),
        'Temperature': [25 + i % 5 + (i * 0.1) for i in range(100)],
        'Humidity': [70 - i % 3 + (i * 0.2) for i in range(100)],
        'Rainfall': [5 + i % 10 - (i * 0.05) for i in range(100)],
        'Patient_Numbers': [100 + i % 20 + (i * 0.5) for i in range(100)]
    }
    df = pd.DataFrame(data)
    print("Created a dummy DataFrame for demonstration.")

# Display the first few rows of the DataFrame
print("\nFirst 5 rows of the DataFrame:")
print(df.head())

# Display data types and non-null values
print("\nDataFrame Info:")
df.info()

# Display descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

**Reasoning**:
Based on the `df.info()` output, the dummy DataFrame has no missing values, so no imputation is needed. The next step is to define the target variable 'malaria outbreaks' by creating a binary column based on a threshold of 'Patient_Numbers'. I will also separate features and the target variable.



```python
print("Checking for missing values...")
print(df.isnull().sum())

# Define 'malaria outbreaks' as the target variable
# For demonstration, let's assume an outbreak occurs if Patient_Numbers exceed the 75th percentile.
outbreak_threshold = df['Patient_Numbers'].quantile(0.75)
df['Outbreak'] = (df['Patient_Numbers'] > outbreak_threshold).astype(int)

print(f"\nMalaria outbreak threshold set at: {outbreak_threshold:.2f} patient numbers (75th percentile).")
print("Number of outbreak instances:")
print(df['Outbreak'].value_counts())

# Separate features and target variable
# Environmental factors are features, 'Outbreak' is the target
features = ['Temperature', 'Humidity', 'Rainfall'] # Assuming these are the environmental factors
target = 'Outbreak'

X = df[features]
y = df[target]

print("\nFeatures (X) head:")
print(X.head())
print("\nTarget (y) head:")
print(y.head())
```

**Reasoning**:
The next step is to apply a scaling technique to the feature data to normalize the values. It's crucial to fit the scaler only on the training data to prevent data leakage. After scaling, the data will be split into training, validation, and test sets while maintaining temporal order, which is essential for time-series data.



```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

# Convert features and target to numpy arrays for scaling and splitting
X_np = X.values
y_np = y.values

# Determine split points for temporal data
# Let's use 70% for training, 15% for validation, and 15% for testing
total_samples = len(df)
train_size = int(total_samples * 0.7)
val_size = int(total_samples * 0.15)
test_size = total_samples - train_size - val_size

X_train, y_train = X_np[:train_size], y_np[:train_size]
X_val, y_val = X_np[train_size:train_size + val_size], y_np[train_size:train_size + val_size]
X_test, y_test = X_np[train_size + val_size:], y_np[train_size + val_size:]

print(f"Original data shape: {X_np.shape}")
print(f"Training data shape: {X_train.shape}, {y_train.shape}")
print(f"Validation data shape: {X_val.shape}, {y_val.shape}")
print(f"Test data shape: {X_test.shape}, {y_test.shape}")

# Apply StandardScaler to features
scaler = StandardScaler()

# Fit on training data and transform all sets
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print("\nFeatures scaled successfully.")
print("First 5 rows of scaled X_train:")
print(X_train_scaled[:5])
```

**Reasoning**:
The data has been loaded, preprocessed, scaled, and split into training, validation, and test sets as NumPy arrays. The next step is to convert these NumPy arrays into PyTorch tensors and then create PyTorch DataLoader objects. This is essential for feeding the data into a PyTorch-based recurrent neural network, fulfilling the 'Preprocess the data for PyTorch' part of the subtask.



```python
import torch
from torch.utils.data import TensorDataset, DataLoader

# Convert NumPy arrays to PyTorch tensors
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val_scaled, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

print("NumPy arrays converted to PyTorch tensors.")
print(f"X_train_tensor shape: {X_train_tensor.shape}")
print(f"y_train_tensor shape: {y_train_tensor.shape}")

# Create TensorDataset
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Define batch size
batch_size = 32 # A common batch size

# Create DataLoader
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False) # Shuffle is False for time series
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"\nCreated DataLoader objects with batch size {batch_size}.")
print(f"Number of batches in training loader: {len(train_loader)}")
print(f"Number of batches in validation loader: {len(val_loader)}")
print(f"Number of batches in test loader: {len(test_loader)}")

# Display a sample batch
print("\nSample batch from training loader:")
for features, labels in train_loader:
    print(f"Features batch shape: {features.shape}")
    print(f"Labels batch shape: {labels.shape}")
    break
```

## Prepare Sequential Data

### Subtask:
Transform the preprocessed features and targets into sequences of a defined `sequence_length`, and then create new PyTorch `TensorDataset` and `DataLoader` objects for the training, validation, and test sets.


**Reasoning**:
The next step is to transform the scaled feature and target data into sequences suitable for a recurrent neural network, and then create new PyTorch DataLoader objects from these sequences.



```python
def create_sequences(input_data, target_data, sequence_length):
    xs, ys = [], []
    for i in range(len(input_data) - sequence_length):
        x = input_data[i:(i + sequence_length)]
        # The target for a sequence is the value immediately after the sequence
        y = target_data[i + sequence_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# Define sequence length
sequence_length = 5  # Example: use 5 past days to predict the 6th day

print(f"Creating sequences with sequence_length = {sequence_length}...")

# Create sequences for training, validation, and test sets
X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train, sequence_length)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, y_val, sequence_length)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test, sequence_length)

print(f"X_train_seq shape: {X_train_seq.shape}, y_train_seq shape: {y_train_seq.shape}")
print(f"X_val_seq shape: {X_val_seq.shape}, y_val_seq shape: {y_val_seq.shape}")
print(f"X_test_seq shape: {X_test_seq.shape}, y_test_seq shape: {y_test_seq.shape}")

# Convert sequential NumPy arrays to PyTorch tensors
X_train_seq_tensor = torch.tensor(X_train_seq, dtype=torch.float32)
y_train_seq_tensor = torch.tensor(y_train_seq, dtype=torch.float32)
X_val_seq_tensor = torch.tensor(X_val_seq, dtype=torch.float32)
y_val_seq_tensor = torch.tensor(y_val_seq, dtype=torch.float32)
X_test_seq_tensor = torch.tensor(X_test_seq, dtype=torch.float32)
y_test_seq_tensor = torch.tensor(y_test_seq, dtype=torch.float32)

print("Sequential NumPy arrays converted to PyTorch tensors.")

# Create new TensorDataset objects
train_dataset_seq = TensorDataset(X_train_seq_tensor, y_train_seq_tensor)
val_dataset_seq = TensorDataset(X_val_seq_tensor, y_val_seq_tensor)
test_dataset_seq = TensorDataset(X_test_seq_tensor, y_test_seq_tensor)

# Create new DataLoader objects (using the previously defined batch_size and shuffle=False)
train_loader_seq = DataLoader(train_dataset_seq, batch_size=batch_size, shuffle=False)
val_loader_seq = DataLoader(val_dataset_seq, batch_size=batch_size, shuffle=False)
test_loader_seq = DataLoader(test_dataset_seq, batch_size=batch_size, shuffle=False)

print(f"\nCreated new DataLoader objects with batch size {batch_size} for sequential data.")
print(f"Number of batches in sequential training loader: {len(train_loader_seq)}")
print(f"Number of batches in sequential validation loader: {len(val_loader_seq)}")
print(f"Number of batches in sequential test loader: {len(test_loader_seq)}")

# Display a sample batch from the new sequential training loader
print("\nSample batch from sequential training loader:")
for features_seq, labels_seq in train_loader_seq:
    print(f"Sequential Features batch shape: {features_seq.shape}")
    print(f"Sequential Labels batch shape: {labels_seq.shape}")
    break
```

## Define PyTorch Model Architecture (Adjusted for Sequences)

### Subtask:
Define a PyTorch recurrent neural network architecture for malaria outbreak prediction, ensuring its forward method is adapted to handle sequential input data.


**Reasoning**:
I need to define a PyTorch recurrent neural network architecture named `MalariaPredictor` that inherits from `nn.Module`, including an LSTM layer and a linear output layer, and implement its `forward` method to process sequential input data for binary classification.



```python
import torch.nn as nn

class MalariaPredictor(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(MalariaPredictor, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        # LSTM layer
        # input_size: number of features in input X (e.g., Temperature, Humidity, Rainfall)
        # hidden_size: number of features in the hidden state h
        # num_layers: number of recurrent layers
        # batch_first=True: input and output tensors are provided as (batch, seq, feature)
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

        # Linear layer to map LSTM output to desired output_size (1 for binary classification)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # Initialize hidden state and cell state with zeros
        # h0 and c0 shape: (num_layers, batch_size, hidden_size)
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # Pass input through LSTM layer
        # out: tensor of shape (batch_size, sequence_length, hidden_size)
        # h_n: final hidden state for each element in the sequence (num_layers, batch_size, hidden_size)
        # c_n: final cell state for each element in the sequence (num_layers, batch_size, hidden_size)
        out, (h_n, c_n) = self.lstm(x, (h0, c0))

        # We are interested in the output of the last time step for classification
        # out[:, -1, :] corresponds to the output of the last sequence element for all batches
        out = self.fc(out[:, -1, :])
        return out

print("Defined MalariaPredictor PyTorch model architecture.")
```

## Train PyTorch Model

### Subtask:
Implement the training loop for the PyTorch model with the sequence-based data. This includes defining the loss function (e.g., BCEWithLogitsLoss for binary classification), selecting an optimizer (e.g., Adam), and iterating through epochs with forward and backward passes. Monitor training and validation loss.


**Reasoning**:
The next step is to implement the training loop for the PyTorch model, including instantiating the model, defining the loss function and optimizer, setting up device-agnostic training, and then executing the epoch-based training and validation process as specified in the instructions.



```python
import torch.optim as optim

# 1. Instantiate the MalariaPredictor model
input_size = X_train_seq.shape[2] # Number of features
hidden_size = 64 # Can be tuned
num_layers = 2 # Can be tuned
output_size = 1 # For binary classification

model = MalariaPredictor(input_size, hidden_size, num_layers, output_size)

# 5. Check for GPU availability and move the model to the appropriate device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f"Model moved to {device}")

# 2. Define the loss function
criterion = nn.BCEWithLogitsLoss() # Combines Sigmoid and BCE Loss for numerical stability

# 3. Define the optimizer
learning_rate = 0.001
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 4. Set the number of training epochs
num_epochs = 50 # Can be tuned

print(f"\nStarting training for {num_epochs} epochs...")

# Lists to store loss for plotting/analysis
train_losses = []
val_losses = []

# 6. Implement a training loop
for epoch in range(num_epochs):
    model.train() # Set model to training mode
    current_train_loss = 0.0
    for batch_idx, (features, labels) in enumerate(train_loader_seq):
        # b. Move features and labels to the correct device
        features = features.to(device)
        labels = labels.to(device).unsqueeze(1) # Add a dimension for BCEWithLogitsLoss

        # g. Reset the gradients using optimizer.zero_grad()
        optimizer.zero_grad()

        # c. Perform a forward pass through the model to get predictions
        outputs = model(features)

        # d. Calculate the loss using the defined loss function
        loss = criterion(outputs, labels)

        # e. Perform a backward pass (backpropagation)
        loss.backward()

        # f. Update the model's weights
        optimizer.step()

        current_train_loss += loss.item()

    avg_train_loss = current_train_loss / len(train_loader_seq)
    train_losses.append(avg_train_loss)

    # 7. Perform a validation step
    model.eval() # Set model to evaluation mode
    current_val_loss = 0.0
    with torch.no_grad(): # b. Disable gradient calculations
        for batch_idx, (features, labels) in enumerate(val_loader_seq):
            # d. Move features and labels to the correct device
            features = features.to(device)
            labels = labels.to(device).unsqueeze(1)

            # e. Get predictions and calculate the validation loss
            outputs = model(features)
            loss = criterion(outputs, labels)
            current_val_loss += loss.item()

    avg_val_loss = current_val_loss / len(val_loader_seq)
    val_losses.append(avg_val_loss)

    # h. Calculate and print the average training and validation loss for the epoch
    if (epoch + 1) % 10 == 0 or epoch == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")

print("\nTraining complete!")
```

Evaluate the trained PyTorch model on the test set. Calculate relevant metrics such as accuracy, precision, recall, and F1-score for classification tasks to assess the model's predictive capability for malaria outbreaks.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

# Convert features and target to numpy arrays for scaling and splitting
X_np = X.values
y_np = y.values

# Determine split points for temporal data
# Let's use 70% for training, 15% for validation, and 15% for testing
total_samples = len(df)
train_size = int(total_samples * 0.7)
val_size = int(total_samples * 0.15)
test_size = total_samples - train_size - val_size

X_train, y_train = X_np[:train_size], y_np[:train_size]
X_val, y_val = X_np[train_size:train_size + val_size], y_np[train_size:train_size + val_size]
X_test, y_test = X_np[train_size + val_size:], y_np[train_size + val_size:]

print(f"Original data shape: {X_np.shape}")
print(f"Training data shape: {X_train.shape}, {y_train.shape}")
print(f"Validation data shape: {X_val.shape}, {y_val.shape}")
print(f"Test data shape: {X_test.shape}, {y_test.shape}")

# Apply StandardScaler to features
scaler = StandardScaler()

# Fit on training data and transform all sets
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print("\nFeatures scaled successfully.")
print("First 5 rows of scaled X_train:")
print(X_train_scaled[:5])
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

# Convert NumPy arrays to PyTorch tensors
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val_scaled, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

print("NumPy arrays converted to PyTorch tensors.")
print(f"X_train_tensor shape: {X_train_tensor.shape}")
print(f"y_train_tensor shape: {y_train_tensor.shape}")

# Create TensorDataset
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Define batch size
batch_size = 32 # A common batch size

# Create DataLoader
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False) # Shuffle is False for time series
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"\nCreated DataLoader objects with batch size {batch_size}.")
print(f"Number of batches in training loader: {len(train_loader)}")
print(f"Number of batches in validation loader: {len(val_loader)}")
print(f"Number of batches in test loader: {len(test_loader)}")

# Display a sample batch
print("\nSample batch from training loader:")
for features, labels in train_loader:
    print(f"Features batch shape: {features.shape}")
    print(f"Labels batch shape: {labels.shape}")
    break
```

```python
def create_sequences(input_data, target_data, sequence_length):
    xs, ys = [], []
    for i in range(len(input_data) - sequence_length):
        x = input_data[i:(i + sequence_length)]
        # The target for a sequence is the value immediately after the sequence
        y = target_data[i + sequence_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# Define sequence length
sequence_length = 5  # Example: use 5 past days to predict the 6th day

print(f"Creating sequences with sequence_length = {sequence_length}...")

# Create sequences for training, validation, and test sets
X_train_seq, y_train_seq = create_sequences(X_train_scaled, y_train, sequence_length)
X_val_seq, y_val_seq = create_sequences(X_val_scaled, y_val, sequence_length)
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test, sequence_length)

print(f"X_train_seq shape: {X_train_seq.shape}, y_train_seq shape: {y_train_seq.shape}")
print(f"X_val_seq shape: {X_val_seq.shape}, y_val_seq shape: {y_val_seq.shape}")
print(f"X_test_seq shape: {X_test_seq.shape}, y_test_seq shape: {y_test_seq.shape}")

# Convert sequential NumPy arrays to PyTorch tensors
X_train_seq_tensor = torch.tensor(X_train_seq, dtype=torch.float32)
y_train_seq_tensor = torch.tensor(y_train_seq, dtype=torch.float32)
X_val_seq_tensor = torch.tensor(X_val_seq, dtype=torch.float32)
y_val_seq_tensor = torch.tensor(y_val_seq, dtype=torch.float32)
X_test_seq_tensor = torch.tensor(X_test_seq, dtype=torch.float32)
y_test_seq_tensor = torch.tensor(y_test_seq, dtype=torch.float32)

print("Sequential NumPy arrays converted to PyTorch tensors.")

# Create new TensorDataset objects
train_dataset_seq = TensorDataset(X_train_seq_tensor, y_train_seq_tensor)
val_dataset_seq = TensorDataset(X_val_seq_tensor, y_val_seq_tensor)
test_dataset_seq = TensorDataset(X_test_seq_tensor, y_test_seq_tensor)

# Create new DataLoader objects (using the previously defined batch_size and shuffle=False)
train_loader_seq = DataLoader(train_dataset_seq, batch_size=batch_size, shuffle=False)
val_loader_seq = DataLoader(val_dataset_seq, batch_size=batch_size, shuffle=False)
test_loader_seq = DataLoader(test_dataset_seq, batch_size=batch_size, shuffle=False)

print(f"\nCreated new DataLoader objects with batch size {batch_size} for sequential data.")
print(f"Number of batches in sequential training loader: {len(train_loader_seq)}")
print(f"Number of batches in sequential validation loader: {len(val_loader_seq)}")
print(f"Number of batches in sequential test loader: {len(test_loader_seq)}")

# Display a sample batch from the new sequential training loader
print("\nSample batch from sequential training loader:")
for features_seq, labels_seq in train_loader_seq:
    print(f"Sequential Features batch shape: {features_seq.shape}")
    print(f"Sequential Labels batch shape: {labels_seq.shape}")
    break
```

```python
import torch.nn as nn

class MalariaPredictor(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(MalariaPredictor, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        # LSTM layer
        # input_size: number of features in input X (e.g., Temperature, Humidity, Rainfall)
        # hidden_size: number of features in the hidden state h
        # num_layers: number of recurrent layers
        # batch_first=True: input and output tensors are provided as (batch, seq, feature)
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

        # Linear layer to map LSTM output to desired output_size (1 for binary classification)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # Initialize hidden state and cell state with zeros
        # h0 and c0 shape: (num_layers, batch_size, hidden_size)
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # Pass input through LSTM layer
        # out: tensor of shape (batch_size, sequence_length, hidden_size)
        # h_n: final hidden state for each element in the sequence (num_layers, batch_size, hidden_size)
        # c_n: final cell state for each element in the sequence (num_layers, batch_size, hidden_size)
        out, (h_n, c_n) = self.lstm(x, (h0, c0))

        # We are interested in the output of the last time step for classification
        # out[:, -1, :] corresponds to the output of the last sequence element for all batches
        out = self.fc(out[:, -1, :])
        return out

print("Defined MalariaPredictor PyTorch model architecture.")
```

```python
import torch.optim as optim

# 1. Instantiate the MalariaPredictor model
input_size = X_train_seq.shape[2] # Number of features
hidden_size = 64 # Can be tuned
num_layers = 2 # Can be tuned
output_size = 1 # For binary classification

model = MalariaPredictor(input_size, hidden_size, num_layers, output_size)

# 5. Check for GPU availability and move the model to the appropriate device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f"Model moved to {device}")

# 2. Define the loss function
criterion = nn.BCEWithLogitsLoss() # Combines Sigmoid and BCE Loss for numerical stability

# 3. Define the optimizer
learning_rate = 0.001
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 4. Set the number of training epochs
num_epochs = 50 # Can be tuned

print(f"\nStarting training for {num_epochs} epochs...")

# Lists to store loss for plotting/analysis
train_losses = []
val_losses = []

# 6. Implement a training loop
for epoch in range(num_epochs):
    model.train() # Set model to training mode
    current_train_loss = 0.0
    for batch_idx, (features, labels) in enumerate(train_loader_seq):
        # b. Move features and labels to the correct device
        features = features.to(device)
        labels = labels.to(device).unsqueeze(1) # Add a dimension for BCEWithLogitsLoss

        # g. Reset the gradients using optimizer.zero_grad()
        optimizer.zero_grad()

        # c. Perform a forward pass through the model to get predictions
        outputs = model(features)

        # d. Calculate the loss using the defined loss function
        loss = criterion(outputs, labels)

        # e. Perform a backward pass (backpropagation)
        loss.backward()

        # f. Update the model's weights
        optimizer.step()

        current_train_loss += loss.item()

    avg_train_loss = current_train_loss / len(train_loader_seq)
    train_losses.append(avg_train_loss)

    # 7. Perform a validation step
    model.eval() # Set model to evaluation mode
    current_val_loss = 0.0
    with torch.no_grad(): # b. Disable gradient calculations
        for batch_idx, (features, labels) in enumerate(val_loader_seq):
            # d. Move features and labels to the correct device
            features = features.to(device)
            labels = labels.to(device).unsqueeze(1)

            # e. Get predictions and calculate the validation loss
            outputs = model(features)
            loss = criterion(outputs, labels)
            current_val_loss += loss.item()

    avg_val_loss = current_val_loss / len(val_loader_seq)
    val_losses.append(avg_val_loss)

    # h. Calculate and print the average training and validation loss for the epoch
    if (epoch + 1) % 10 == 0 or epoch == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")

print("\nTraining complete!")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("\nEvaluating model performance on the test set...")

# 1. Set the model to evaluation mode
model.eval()

# 2. Initialize empty lists to store all true labels and predicted probabilities/labels
all_labels = []
all_predictions = []

# 3. Iterate through the test_loader_seq without gradient calculations
with torch.no_grad():
    for features, labels in test_loader_seq:
        # 4. Move features and labels to the correct device
        features = features.to(device)
        labels = labels.to(device)

        # 5. Get the model's outputs (logits)
        outputs = model(features)

        # 6. Apply a sigmoid activation to the logits to get predicted probabilities
        probabilities = torch.sigmoid(outputs)

        # 7. Convert the predicted probabilities to binary predictions using a threshold (e.g., 0.5)
        predictions = (probabilities > 0.5).float()

        # 8. Store the true labels and predicted labels
        all_labels.extend(labels.cpu().numpy())
        all_predictions.extend(predictions.cpu().numpy())

# 9. Convert the stored true labels and predicted labels to NumPy arrays
all_labels = np.array(all_labels)
all_predictions = np.array(all_predictions).flatten() # Flatten in case predictions have shape (N,1)

# 10. Calculate and print classification metrics
accuracy = accuracy_score(all_labels, all_predictions)
precision = precision_score(all_labels, all_predictions, average='binary', zero_division=0)
recall = recall_score(all_labels, all_predictions, average='binary', zero_division=0)
f1 = f1_score(all_labels, all_predictions, average='binary', zero_division=0)

print(f"\nAccuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

print("Model evaluation complete.")
```

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

print("Generating ROC curve and Confusion Matrix...")

# Re-initialize lists to store all true labels and predicted probabilities
all_labels = []
all_probabilities = []

# Set the model to evaluation mode
model.eval()

# Iterate through the test_loader_seq without gradient calculations
with torch.no_grad():
    for features, labels in test_loader_seq:
        # Move features and labels to the correct device
        features = features.to(device)
        labels = labels.to(device)

        # Get the model's outputs (logits)
        outputs = model(features)

        # Apply a sigmoid activation to the logits to get predicted probabilities
        probabilities = torch.sigmoid(outputs)

        # Store the true labels and predicted probabilities
        all_labels.extend(labels.cpu().numpy())
        all_probabilities.extend(probabilities.cpu().numpy())

# Convert the stored true labels and predicted probabilities to NumPy arrays
all_labels = np.array(all_labels)
all_probabilities = np.array(all_probabilities).flatten()

# --- Generate ROC Curve ---

# Check if there are both positive and negative samples in all_labels
if len(np.unique(all_labels)) < 2:
    print("\nSkipping ROC curve generation: Test set contains only one class in true labels.")
    roc_auc = np.nan # Set ROC AUC to NaN as it cannot be calculated
else:
    fpr, tpr, thresholds = roc_curve(all_labels, all_probabilities)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Chance')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

print(f"ROC AUC: {roc_auc:.4f}")

# --- Generate Confusion Matrix ---

# Convert predicted probabilities to binary predictions using a threshold (e.g., 0.5)
binary_predictions = (all_probabilities > 0.5).astype(int)

cm = confusion_matrix(all_labels, binary_predictions)

plt.figure(figsize=(8, 6))
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Outbreak', 'Outbreak'])
display.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

print("ROC curve and Confusion Matrix generated successfully.")
```
